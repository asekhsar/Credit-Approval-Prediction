### Importing the libraries:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

### Loading the dataset:

```python
header_list=['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10','A11','A12','A13','A14','A15','A16']
df=pd.read_csv('crx.data',names=header_list)
pd.set_option('display.max_rows', 1000)
new_df=df.replace(to_replace ='?', value =np.nan)
new_df
```

### Checking for null values:

```python
new_df.isnull()
new_df.info()
new_df
```

```python
new_df_1=new_df.dropna()
```

```python
new_df_1.info()
```

```python
new_df_1
```

### Treating the categorical and continous features seperately:

```python
new_df_ca=new_df_1[['A1','A4','A5','A6','A7','A9','A10','A12','A13']]
```

```python
new_df_ca
```

```python
new_df_cp=new_df_1[['A2','A3','A8','A11','A14','A15']]
new_df_cp["A2"] = pd.to_numeric(new_df_cp["A2"], downcast="float")
new_df_cp["A3"] = pd.to_numeric(new_df_cp["A3"], downcast="float")
new_df_cp["A8"] = pd.to_numeric(new_df_cp["A8"], downcast="float")
new_df_cp["A11"] = pd.to_numeric(new_df_cp["A11"], downcast="float")
new_df_cp["A14"] = pd.to_numeric(new_df_cp["A14"], downcast="float")
new_df_cp["A15"] = pd.to_numeric(new_df_cp["A15"], downcast="float")
```

```python
new_df_cp
```

```python
new_df_target=new_df_1[['A16']]
```

```python

value_dict={}
for col in new_df_ca.columns:
        unique_values=new_df_ca[col].unique()
        a=0
        if col not in value_dict:
            value_dict[col]={}
        
        for value in sorted(unique_values):
            new_df_ca[col].replace(value,a,inplace=True)
           
            if value not in value_dict[col]:
                value_dict[col][value]=a
            a+=1
            
      
        
        
          
        
```

```python
value_dict
```

```python
new_df_ca
```

```python
for column in new_df_cp.columns:
        new_df_cp[column] = (new_df_cp[column] - new_df_cp[column].mean()) / new_df_cp[column].std()
```

```python
new_df_cp
```

```python
df_final=[new_df_ca,new_df_cp]
final=pd.concat(df_final,axis=1)
```

### Final DataFrame to be fed into the model:

```python
final
```

```python
unique_values=new_df_target['A16'].unique()
b=0
for value in sorted(unique_values):
    new_df_target['A16'].replace(value,b,inplace=True)
    b+=1
```

```python
new_df_target
```

```python
(new_df_target['A16'] == 0).sum()
```

```python
(new_df_target['A16'] == 1).sum()
```

### Performing the train test split for the training and testing dataset:

```python
X_train, X_test, y_train, y_test = train_test_split(final, new_df_target, random_state=0)
```

### KNN

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
y_pred = knn.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
pd.crosstab(np.array(y_test.values.tolist()).reshape((-1,)), y_pred, rownames=['True'], colnames=['Predicted'])
```

```python
np.array(y_test.values.tolist()).reshape((-1,))
```

### Logistic Regression

```python
model = LogisticRegression(solver='liblinear', random_state=0)
model = LogisticRegression(solver='liblinear', random_state=0).fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### SVM

```python
svclassifier = SVC(kernel='linear')
```

```python
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```
